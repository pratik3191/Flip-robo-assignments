###  Write a python program to scrap book name, author name, genre and book review of any 5 books from
### ‘www.bookpage.com’


```python
from bs4 import BeautifulSoup
import requests
```

```python
page=requests.get("https://bookpage.com/reviews")
```

```python
page
```

```python
soup=BeautifulSoup(page.content,'html.parser')
soup
```

```python
Books=soup.find_all('div',class_="flex-article-content")
```

```python
Books
```

```python
firstbook=Books[0]
```

```python
firstbook.find('p',class_="excerpt").text.replace('\n','')
```

```python
book_title=[]
Author=[]
Genre=[]
review=[]
for i in Books:
    book_title.append(i.h4.a.text[3:])
    Author.append(i.find('p',class_="sans bold").text[1:26])
    Genre.append(i.find('p',class_="genre-links hidden-phone").text.replace('\n',''))
    review.append(i.find('p',class_="excerpt").text.replace('\n',''))
```

```python
data=list(zip(book_title,Author,Genre,review))
```

```python
import pandas as pd
df=pd.DataFrame(data,columns=['Book Name','Author','Genre','Book Review'])
```

```python
df.to_csv("bookreview.csv")
```
