### Write a python program to display all the header tags from ‘en.wikipedia.org/wiki/Main_Page’.

```python
from bs4 import BeautifulSoup
import requests
```

```python
page=requests.get("https://en.wikipedia.org/wiki/Main_Page")
```

```python
page
```

```python
soup=BeautifulSoup(page.content,'html.parser')
soup
```

```python
heading1=soup.find_all('h1')
```

```python
headingtag=[]
for i in heading1:
    headingtag.append(i.text.replace('\n',''))
headingtag    
```

```python
heading2=soup.find_all('h2')
```

```python
for i in heading2:
    headingtag.append(i.text.replace('\n',''))
headingtag    
```

```python
heading3=soup.find_all('h3')
```
